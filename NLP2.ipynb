```python
from gensim import corpora

# Sample documents
documents = [
    "The quick brown fox jumps over the lazy dog",
    "I love machine learning and deep learning",
    "Gensim is a great library for NLP tasks"
]

# Tokenize the documents
tokenized_docs = [doc.lower().split() for doc in documents]

# Create a dictionary from the tokenized documents
my_dictionary = corpora.Dictionary(tokenized_docs)

# Create the BoW corpus (list of lists of (token_id, token_count) tuples)
BoW_corpus = [my_dictionary.doc2bow(doc) for doc in tokenized_docs]

# Initialize an empty list to store word weights
word_weight = []

# Iterate through the corpus to extract word frequencies
for doc in BoW_corpus:
    for id, freq in doc:
        word_weight.append([my_dictionary[id], freq])

# Print the word weights
print(word_weight)
```

```python
import gensim.downloader as api
from gensim.models.phrases import Phrases

# Load the text8 dataset
dataset = api.load("text8")

# Extract a list of sentences from the dataset
data = []
for sentence in dataset:
    data.append(sentence)  # This line must be indented properly

# Bigram using Phrases Model             
bigram_model = Phrases(data, min_count=3, threshold=10)

# Apply the bigram model to the first sentence
print(bigram_model[data[0]])
```

```python
import gensim.downloader as api
from multiprocessing import cpu_count
from gensim.models.word2vec import Word2Vec

# load the text8 dataset
dataset = api.load("text8")

# extract a list of sentences from the dataset
data = []
for sentence in dataset:
    data.append(sentence)  # Proper indentation here

# We will split the data into two parts
data_1 = data[:1200]  # This is used to train the model
data_2 = data[1200:]  # This part will be used to update the model

# Training the Word2Vec model
w2v_model = Word2Vec(data_1, min_count=0, workers=cpu_count())

# Word vector for the word "time"
print(w2v_model.wv['time'])  # Use .wv to access word vectors
```
